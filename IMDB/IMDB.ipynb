```python
genre_rating = dict()
genre_total = dict()
genre_rating_average = dict()
for film in data.iterrows():
    if(math.isnan(film[1]['imdb_score'])):
        continue
    genres = film[1]["genres"].split('|')
    for genre in genres:
        if(genre_rating.has_key(genre) == False):
            genre_rating[genre] = film[1]['imdb_score']
            genre_total[genre] = 1
            genre_rating_average[genre] = film[1]['imdb_score']
            continue 
        genre_rating[genre] += film[1]['imdb_score']
        genre_total[genre] +=1
        genre_rating_average[genre] =  genre_rating[genre] \
                                                            / genre_total[genre]
    
```
